```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, metrics
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor  # Use RandomForestClassifier if your target is categorical
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
#import data

batting_df = pd.read_csv("../Data/teams_batting_00-24.csv").drop("Unnamed: 0",axis=1)
batting_df =batting_df[batting_df["Tm"]!="League Average"]
batting_df.tail()
```

```python
#We extrat the desired variables
y=batting_df["R"]
x = batting_df[batting_df.columns[9:12].append(batting_df.columns[13:-1])]
x.head()
```

```python
X_train, X_test,y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
#We create the model

reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
# regression coefficients
print('Coefficients: ', reg.coef_)

# variance score: 1 means perfect prediction
print('Variance score: {}'.format(reg.score(X_test, y_test)))
```

```python
results_dict = {"Variable": x.columns,
               "Coeficiente":reg.coef_}
results_df = pd.DataFrame(results_dict)
results_df
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

y_pred=reg.predict(X_test)
mean_absolute_percentage_error(y_test, y_pred)
```

```python
teams_dictionary={'New York Mets':'NYM',
                 'Los Angeles Dodgers':'LAD',
                 'Toronto Blue Jays':'TOR',
                 'Texas Rangers':'TEX',
                 'St. Louis Cardinals':'STL',
                 'San Francisco Giants':'SFG',
                 'Seattle Mariners':'SEA',
                 'San Diego Padres':'SDP',
                 'Pittsburgh Pirates':'PIT',
                 'Philadelphia Phillies':'PHI',
                 'Oakland Athletics':'OAK',
                 'New York Yankees':'NYY',
                 'Arizona Diamondbacks':'ARI',
                 'Milwaukee Brewers':'MIL',
                 'Minnesota Twins':'MIN',
                 'Kansas City Royals':'KCR',
                 'Cincinnati Reds':'CIN',
                 'Atlanta Braves':'ATL',
                 'Baltimore Orioles':'BAL',
                 'Boston Red Sox':'BOS',
                 'Chicago Cubs':'CHC',
                 'Chicago White Sox':'CHW',
                 'Colorado Rockies':'COL',
                 'Detroit Tigers':'DET',
                 'Houston Astros':'HOU',
                 'Cleveland Indians':'CLE',
                 'Washington Nationals':'WSN',
                 'Tampa Bay Rays':'TBR',
                 'Miami Marlins':'MIA',
                 'Florida Marlins':'FLA',
                 'Los Angeles Angels of Anaheim':'ANA',
                 'Anaheim Angels':'ANA',
                 'Montreal Expos':'MON',
                 'Cleveland Guardians':'CLE',
                 'Los Angeles Angels':'LAA',
                 'Tampa Bay Devil Rays':'TBD'}

batting_df["Team"] = batting_df["Team"].replace(teams_dictionary)
```

```python
merged_df = pd.merge(batting_df, advance_df, on=["Team","Year"], how="inner")
merged_df
```

```python
model_df = merged_df[["Team", "R", "HR", "OBP","SLG","SB","RBI","BA","W rate", "K rate", "BABIP","wOBA"]]
```

```python
model_df["ISO"] = model_df["SLG"] - model_df["BA"]
model_df
```

```python
# Define the target column. Replace 'Runs' with your actual target column.
target_column = 'R'

# Prepare the features by dropping columns that you don't want to include.
# Here we drop 'Team' (an identifier) and the target column from the features.
features = model_df.drop(columns=['Team', target_column,"RBI"])
target = model_df[target_column]
```

```python
# Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create the Random Forest model.
# For a regression task, we use RandomForestRegressor.
# (If you have a classification task, use RandomForestClassifier instead.)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions on the test set.
y_pred = rf_model.predict(X_test)

# Evaluate the model. Here, we calculate the Mean Squared Error (MSE).
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on test set:", mse)
mape = mean_absolute_percentage_error(y_test, y_pred)
print("Mean Squared Error on test set:", mape)

# Optionally, display feature importances:
importances = rf_model.feature_importances_
feature_names = features.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
print(feature_importance_df.sort_values(by='Importance', ascending=False))
```
